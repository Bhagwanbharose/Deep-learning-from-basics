```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

### Sigmoid Function!

```python
def sigmoid(x):
    return (1 / (1 + np.exp(-x)))
```

```python
data = np.linspace(-5,5,100)
plt.plot(data)
```

```python
plt.plot(sigmoid(data))
```

### Tanh Function!

```python
def tanh(x):
    return((np.exp(+x)-np.exp(-x))/((np.exp(+x)+np.exp(-x))))
```

```python
plt.plot(tanh(data))
```
