```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
data = np.linspace(0,10,10) + np.random.uniform(-1,1,10)
data
```

```python
plt.plot(data)
```

```python
label = np.linspace(0,10,10) + np.random.uniform(-1,1,10)
```

```python
plt.plot(label)
```

```python
plt.plot(data,label,'*')
```

### y = m * x + c

```python
np.random.rand(2)
```

```python
m = tf.Variable(0.90)
c = tf.Variable(0.43)
```

```python
error = 0

for x,y in zip(data,label):
    pred_label = m*x+c
    error = error + (label-pred_label)**2 
```

```python
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.001)
train = optimizer.minimize(error)
```

```python
initialize_all = tf.global_variables_initializer()
```

```python
with tf.Session() as session:
    session.run(initialize_all)
    training_steps = 10
    for i in range(training_steps):
        session.run(train)
    slope , intercept  = session.run([m,c])
```

```python
x_test = np.linspace(-1,10,10)

y_pred = slope*x_test+intercept
```

```python
plt.plot(data,label,'*')
plt.plot(x_test,y_pred,'r')
# plt.figure()
```
